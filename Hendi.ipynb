```python

import cv2
import numpy as np

img = cv2.imread('images/images.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
img = cv2.imread('images/STOP.png',0)
rows,cols = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))

final_frame = cv2.hconcat((img, dst))

cv2.imshow('image',final_frame)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
img = cv2.imread('images/samsung.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[300,50],[50,300]])
pts2 = np.float32([[10,100],[300,50],[100,350]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

Thresholds

```python
# import opencv
import cv2
import numpy as np
# Read image

from skimage.filters import threshold_mean


image =  cv2.imread("images/thesholds.jpg", cv2.IMREAD_GRAYSCALE)

blur = cv2.blur(image,(6,6))
thresh = threshold_mean(blur)

binary = blur > thresh




fig, axes = plt.subplots(ncols=3, figsize=(8, 3))
ax = axes.ravel()

ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].set_title('Original image')

ax[1].imshow(image, cmap=plt.cm.gray)
ax[1].set_title('Blur image')

ax[2].imshow(binary, cmap=plt.cm.gray)
ax[2
  ].set_title('Result')



for a in ax:
    a.axis('off')

plt.show()
```

# Sharpiningv TEST

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
 
img = cv2.imread('images/tree.jpg')
 
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
im = cv2.filter2D(img, -1, kernel) 
cv2.imshow('Normal', img);# to display image (shift+enter).

cv2.imshow('shapining', im);# to display image (shift+enter).

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(im),plt.title('Sharpining')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('images/tree.jpg',0)

laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
final = cv2.add(sobelx,sobely)
final2 = cv2.add(final,laplacian)

plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
cv2.imshow('1',final)
cv2.imshow('2',final2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
